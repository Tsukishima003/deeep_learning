**importing libraries**

```python
import numpy as np
from numpy import array 
from numpy import random 
from numpy import dot 
from random import choice
```

***Intialized dataset***

```python
dataset = [ 
(array([0,0,1]), 0), 
(array([0,1,1]), 1), 
(array([1,0,1]), 1), 
(array([1,1,1]), 1), 
] 
print(dataset)
print(array([0,0,1]),0)
```

**initializing random numbers for WEIGHTS**

```python
weights = random.rand(3)
weights
```

**initializing additional variables**

```python
r = 0.4
n = 100
```

**Step Activation Fn**

```python
activation_fn = lambda x: 0 if x < 0 else 1
for j in range(n):
    x , expected  = choice(dataset)
    res = dot(weights , x)
    err = expected - activation_fn(res)
    weights += r * err * x
for x ,_ in dataset:
    res = dot( x , weights)
    print("ResultBAFn : {} ResultAFN:{}".format(round(res,3), activation_fn(res)))
```

**Linear activation Fn**

```python
activaitonFn = lambda x: x
for j in range(n):
    x , expected = choice(dataset)
    resu = dot(weights , x)
    err = expected - activaitonFn(resu)
    weights += r * err * x
for x , _ in dataset:
    resu = dot(x , weights)
    print("ResultBAFn : {} ResultAFN:{}".format(round(res,3), activation_fn(res)))
```

**Sigmoid Activatoin fn**

```python
import numpy as np 
activationFn = lambda x: 1/(1+np.exp(-x)) 
for j in range(n): 
    x, expected = choice (dataset) 
    result = dot (weights, x) 
    err = expected-activationFn(result)
    weights += r * err * x 
for x, _ in dataset:  
    result = dot(x, weights) 
    print("ResultBAFn: {} ResultAFn: {}".format(round (result,3), activationFn(result))) 
```

**ReLu Actiavton fn**

```python
activationFn = lambda x: 0 if x < 0 else x 
for j in range(n): 
    x, expected = choice(dataset) 
    result = dot(weights, x) 
    err = expected-activationFn(result)
    weights += r * err * x 
for x, _  in dataset: 
    result = dot(x, weights) 
    print("ResultBAFn: {} ResultAFn: {}".format(round (result,3), activationFn(result)))
```
